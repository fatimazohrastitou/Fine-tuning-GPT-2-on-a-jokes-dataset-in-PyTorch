```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, GRU, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Exemple de textes
texts = ["هذا نص باللغة العربية", "النص الثاني باللغة العربية"]

# Création et ajustement du tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

# Taille du vocabulaire
vocab_size = len(tokenizer.word_index) + 1  # Ajouter 1 pour inclure l'indice de padding

# Convertir les textes en séquences et ajouter du padding
sequences = tokenizer.texts_to_sequences(texts)
max_length = 10  # Longueur maximale des séquences
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')

# Exemple de scores (labels)
scores = [6, 7.5]

# Normalisation des scores entre 0 et 1 (par exemple)
normalized_scores = np.array(scores) / 10.0

# Définir le modèle RNN
model_rnn = Sequential([
    Embedding(input_dim=vocab_size, output_dim=64, input_length=max_length),
    SimpleRNN(64, activation='tanh'),
    Dense(1, activation='sigmoid')
])

# Compiler le modèle RNN
model_rnn.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Définir le modèle GRU
model_gru = Sequential([
    Embedding(input_dim=vocab_size, output_dim=64, input_length=max_length),
    GRU(64, activation='tanh'),
    Dense(1, activation='sigmoid')
])

# Compiler le modèle GRU
model_gru.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Définir le modèle LSTM
model_lstm = Sequential([
    Embedding(input_dim=vocab_size, output_dim=64, input_length=max_length),
    LSTM(64, activation='tanh'),
    Dense(1, activation='sigmoid')
])

# Compiler le modèle LSTM
model_lstm.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Entraîner les modèles
model_rnn.fit(padded_sequences, normalized_scores, epochs=10, batch_size=2)
model_gru.fit(padded_sequences, normalized_scores, epochs=10, batch_size=2)
model_lstm.fit(padded_sequences, normalized_scores, epochs=10, batch_size=2)

# Évaluer les modèles
loss_rnn, mae_rnn = model_rnn.evaluate(padded_sequences, normalized_scores)
print(f"RNN Model - Loss: {loss_rnn}, MAE: {mae_rnn}")

loss_gru, mae_gru = model_gru.evaluate(padded_sequences, normalized_scores)
print(f"GRU Model - Loss: {loss_gru}, MAE: {mae_gru}")

loss_lstm, mae_lstm = model_lstm.evaluate(padded_sequences, normalized_scores)
print(f"LSTM Model - Loss: {loss_lstm}, MAE: {mae_lstm}")
```

```python
import keras_tuner as kt

def build_model(hp):
    model = Sequential([
        Embedding(
            input_dim=vocab_size,
            output_dim=hp.Int('embedding_dim', min_value=32, max_value=128, step=32),
            input_length=max_length
        ),
        LSTM(hp.Int('units', min_value=32, max_value=128, step=32)),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

tuner = kt.Hyperband(
    build_model,
    objective='val_mae',
    max_epochs=20,
    directory='tuner_dir',
    project_name='text_model'
)

tuner.search(padded_sequences, normalized_scores, validation_split=0.2, epochs=20, batch_size=16)
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import sacrebleu

# Example data: 
# For classification tasks
Y_true = [1, 0, 1, 1, 0]  # True labels
Y_pred = [1, 0, 1, 0, 0]  # Predicted labels

# For text generation tasks (BLEU score)
predictions = ["this is a test", "hello world"]
references = ["this is a test", "hello world"]

# Classification Metrics
accuracy = accuracy_score(Y_true, Y_pred)
precision = precision_score(Y_true, Y_pred, average='binary')
recall = recall_score(Y_true, Y_pred, average='binary')
f1 = f1_score(Y_true, Y_pred, average='binary')

print(f"Classification Metrics:")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")

# BLEU Score
bleu_score = sacrebleu.corpus_bleu(predictions, [references])
print(f"BLEU Score: {bleu_score.score}")
```

```python
from datasets import load_dataset
from transformers import Trainer, TrainingArguments, GPT2Tokenizer, GPT2LMHeadModel

# Load the dataset
dataset = load_dataset("text", data_files={"train": "custom_dataset.txt"})

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, padding="max_length", max_length=512)

tokenized_datasets = dataset.map(tokenize_function, batched=True, num_proc=4, remove_columns=["text"])

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    save_steps=10_000,
    save_total_limit=2,
    learning_rate=5e-5,
    logging_dir='./logs',
)

# Fine-tune the model
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
)

trainer.train()
```
